```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

url = 'https://www.bbq.co.kr/menu/menuList.asp'
data = req.urlopen(url)
soup = BeautifulSoup(data, 'html.parser')

name = soup.select('div.info > p.name')
names = []
for i in name :
    names.append(i.string)
    
price = soup.select('div.info > p.pay')
prices = []
for i in price :
    prices.append(i.string)

header = ['메뉴', '가격']
df = pd.DataFrame([x for x in zip(names,prices)], columns=header)
print(df)

# 문자열인 가격
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

url = 'https://www.bbq.co.kr/menu/menuList.asp'
data = req.urlopen(url)
soup = BeautifulSoup(data, 'html.parser')

name = soup.select('div.info > p.name')
names = []
for i in name :
    names.append(i.string)
    
price = soup.select('div.info > p.pay')
prices = []
for i in price :
    prices.append(i.string)

header = ['메뉴', '가격']
df = pd.DataFrame([x for x in zip(names,prices)], columns=header)
print(df)
```
